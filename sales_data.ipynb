```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="sales_data",
    user="postgres",
    password="Loganathan@456",
    port="5432"
)

cursor = conn.cursor()

print("Connected successfully!")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales_data (
    Order_ID VARCHAR(20),
    Order_Date DATE,
    Customer_Name VARCHAR(100),
    Category VARCHAR(50),
    Product_Name VARCHAR(100),
    Quantity INTEGER,
    Unit_Price NUMERIC(10,2),
    Sales NUMERIC(12,2),
    Profit NUMERIC(12,2),
    Region VARCHAR(50),
    City VARCHAR(100),
    Payment_Method VARCHAR(50),
    Rating NUMERIC(3,1)
)
""")

conn.commit()

print("Table created successfully!")
```
